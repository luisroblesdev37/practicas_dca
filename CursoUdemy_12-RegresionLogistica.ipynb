# La Regresion Logistica

#### permite estimar la probabilidad de una variable cualitativa binaria, en funcion de una variable cuantitativa

#### es decir permite la clasificacion binaria, solo 2 valores o dicotomica.

```python
import numpy as np
import matplotlib.pyplot as plt
import math
x = np.linspace(-10,10)
y = 0 +1.5*x

plt.plot(x,y,'r')
plt.title('regresion lineal')
plt.show()

def function(data):
    return 1/(1 + math.e**(-1.5*data))

y1 = function(x)
plt.plot(x, y1,'r')
plt.title('regresion logistica')
plt.show()
```
